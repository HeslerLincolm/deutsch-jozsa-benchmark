# Boolean Functions for the Deutsch–Jozsa Algorithm

In this section, we generate Boolean functions that will be used to test the classical and quantum algorithms.

A Boolean function is defined as:

$$f: \{0,1\}^n \to \{0,1\}$$

The **Deutsch–Jozsa problem** consists of determining whether a function is:

* **Constant:** $f(x)$ is the same for all $x$.
* **Balanced:** $f(x) = 0$ for exactly half of the inputs and $f(x) = 1$ for the other half.

```python
#  Import libraries
import numpy as np
import itertools
```

```python
# Create binary entries
def generate_inputs(n):
    return list(itertools.product([0,1], repeat=n))

inputs = generate_inputs(3)
inputs
```

### Constant function

```python
# Create constant function
def constant_zero(x):
    return 0

def constant_one(x):
    return 1
```

```python
# Example
for x in inputs:
    print(x, constant_one(x))
```

### Balanced function

```python
# Create balanced function
# using XOR
def balanced_xor(x):
    return x[0] ^ x[1]
```

```python
# Example
for x in inputs:
    print(x, balanced_xor(x))
```

```python
# Check if a function is balanced
def check_function(f, inputs):

    outputs = [f(x) for x in inputs]

    zeros = outputs.count(0)
    ones = outputs.count(1)

    if zeros == len(inputs):
        return "constant"

    if ones == len(inputs):
        return "constant"

    if zeros == ones:
        return "balanced"

    return "unknown"
```

```python
# Example
print(check_function(constant_one, inputs))
print(check_function(balanced_xor, inputs))
```
