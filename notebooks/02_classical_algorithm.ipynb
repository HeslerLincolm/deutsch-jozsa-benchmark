### Analyze the classic cost

A classical deterministic algorithm may need to evaluate the function on many inputs in order to determine whether it is constant or balanced. In the worst case it may require evaluating half of the possible inputs plus one.

```python
# Number of possible evaluations
for n in range(1,7):

    inputs = generate_inputs(n)

    print("n =", n, "number of inputs =", len(inputs))
```

```python
# Classical algorithm
inputs = generate_inputs(3)

result = classical_deutsch_jozsa(balanced_xor, inputs)

print(result)
```

To guarantee a correct answer, the algorithm requires a worst-case complexity of:
  $$T_{classical} = 2^{n-1} + 1$$
  This demonstrates that classical complexity grows exponentially with the number of bits $n$.

```python
n_values = np.arange(1, 11)

classical_complexity = 2**(n_values-1) + 1

print(n_values)
print(classical_complexity)
```

```python
# Plot
plt.figure()

plt.plot(n_values, classical_complexity, marker="o")

plt.xlabel("Number of bits (n)")
plt.ylabel("Worst-case classical evaluations")
plt.title("Classical Complexity of the Deutsch–Jozsa Problem")

plt.grid(True)

plt.show()
```

This exponential growth highlights the limitations of classical algorithms for certain problems and motivates the study of quantum algorithms such as the Deutsch–Jozsa algorithm.
