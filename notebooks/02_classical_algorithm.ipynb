### Analyze the classic cost

A classical deterministic algorithm may need to evaluate the function on many inputs in order to determine whether it is constant or balanced. In the worst case it may require evaluating half of the possible inputs plus one.

```python
# Number of possible evaluations
for n in range(1,7):

    inputs = generate_inputs(n)

    print("n =", n, "number of inputs =", len(inputs))
```

To guarantee a correct answer, the algorithm requires a worst-case complexity of:
  $$T_{classical} = 2^{n-1} + 1$$
  This demonstrates that classical complexity grows exponentially with the number of bits $n$.
